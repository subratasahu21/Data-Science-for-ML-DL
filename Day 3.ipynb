```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(5, 100, 100)
y=np.linspace(10, 1000, 100)
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot([1,4,6,8],[3,8,3,5])
plt.xlabel("Evan")
plt.ylabel("Moores")
plt.title("Birla")
plt.show()
```

```python
plt.plot(x,y, 'b+')
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("ohm's law")
plt.xlim([20, 80])
plt.ylim([200, 800])
plt.show()
```

```python
x=np.linspace(0,10,20)
y=x*2
plt.plot(x,y, 'b+')
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("ohm's law")
plt.show()
```

```python
x=np.linspace(0,5,10)
y=np.linspace(3,6,10)
plt.plot(x,y,'r-',x,y**2,'b+',x,y**3,'g^')
plt.show()
```

```python
x=np.linspace(1,10,100)
y=np.log(x)
plt.figure(1)

plt.subplot(3,1,1)
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(3,1,2)
plt.title("y=log(x)**2")
plt.plot(x,y**2)

plt.subplot(3,1,3)
plt.title("y=y*2")
plt.plot(x,y*2)
```

```python
x=np.linspace(1,10,100)
plt.figure(1)
plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title("Cube")
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title("Log")
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title("Exponential")
plt.plot(x,x**2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
df=pd.read_csv("market_fact.csv")
```

```python
sns.distplot(df['Shipping_Cost'])
```

```python
sns.distplot(df['Shipping_Cost'][:200], rug=True)
```

```python
sns.distplot(df['Sales'], hist=False)
```

```python
sns.boxplot(y=df['Order_Quantity'])
plt.title('Order Quantity')
```

```python
sns.jointplot(x=df['Sales'], y=df['Profit'])
plt.show()
```

```python
df=df[(df.Profit<10000) & (df.Sales<20000)]
sns.jointplot(x=df['Sales'], y=df['Profit'])
plt.show()
```

```python
df=pd.read_csv('market_fact.csv')
df=df[(df.Profit<100) & (df.Profit>-100) & (df.Sales<200)]
sns.jointplot(x=df['Sales'], y=df['Profit'], kind="hex", color="k")
```
